```python
import time

def gcd_simple(a, b):
    gcd = 1
    for i in range(1, min(a, b) + 1):
        if a % i == 0 and b % i == 0:
            gcd = i
    return gcd


# Input values
a = 123456
b = 789012

# Start time
start_time = time.time()

# GCD computation
result = gcd_simple(a, b)

# End time
end_time = time.time()

# Elapsed time in milliseconds
elapsed_time_ms = (end_time - start_time) * 1000

print("GCD:", result)
print("Elapsed time:", elapsed_time_ms, "ms")

```

```python
# Function that returns the factorization of an integer (ascending approach)
def fact_int(n):
    nn = n**(1/2)
    m = 2

    while m < nn:
        if n % m == 0:
            return (m, n // m)
        m += 1

    return (1, n)
```

```python
import time
import math

def fact_int_2(n):
    m = math.isqrt(n)
    while m > 1:
        if n % m == 0:
            return (m, n // m)
        m -= 1
    return (1, n)


# Test number
n = 1234567890

# Start time
start = time.time()

# Function call
factors = fact_int_2(n)

# End time
end = time.time()

# Elapsed time in milliseconds
elapsed_ms = (end - start) * 1000

print("Factors:", factors)
print("Elapsed time:", elapsed_ms, "ms")
```
